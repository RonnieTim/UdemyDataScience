##### There is an ice cream shop, that is operating in New York, LA and San Francisco.			
##### You have sold 12,327 ice creams in New York; 17,129 in LA and 19,923 in San Francisco.			
##### Order the data in a frequency distribution table.

```python
df = pd.DataFrame({'Ice Cream Sales(Freq)': [12327, 17129, 19923]})
df.index = ['New York', 'LA', 'San Francisco']
df
```

##### You have a frequency distribution table with all the sales.	Create a bar chart, representing the data. Note that in Excel, the term for a bar charts are called both:  'column charts' and 'bar charts'.

```python
df.plot.bar(rot=20, title='IceCream sales')
plot.show(block=True);
```

#### Create a new column in your table, representing the relative frequency of input. You can choose to express it in percentages or as a decimal.	Create a pie chart, representing the share of each city to the sales of your company.	

```python
df_copy = pd.DataFrame({'Ice Cream Sales(Freq)': [12327, 17129, 19923]})
df_copy.index = ['New York', 'LA', 'San Francisco']
df_copy
```

```python
total = sum([12327, 17129, 19923])
total
```

```python
rel1 = 12327 / total
rel1
rel2 = 17129 / total
rel2
rel3 = 19923 / total
relative_freq = [rel1, rel2, rel3]
relative_freq
```

```python
df_copy['Relative Frequency'] = relative_freq
df_copy
```

```python
df_copy['Relative Frequency'].plot.pie(figsize=(5, 5))
plot.show()
```

```python
cum_freq = [rel1, rel1 + rel2, rel1 + rel2 + rel3]
cum_freq
```

```python
df_copy['Cumulative Frequency'] = cum_freq
df_copy
```

##### In a new column, calculate the cumulative frequency of the data. 			On a second axis in the same chart, represent the cumulative frequency of the data.

```python
df_copy = df_copy.sort_values(by='Ice Cream Sales(Freq)',ascending=False)
df_copy['CumPercentage'] = df_copy['Ice Cream Sales(Freq)'].cumsum()/df_copy['Ice Cream Sales(Freq)'].sum()*100
```
